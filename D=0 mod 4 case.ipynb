```python
# parameters: known a_ijk and D
Pa = PolynomialRing(QQ, [
    'a111','a112','a121','a122',
    'a211','a212','a221','a222',
    'D'
])
a111,a112,a121,a122,a211,a212,a221,a222,D = Pa.gens()
Ka = Pa.fraction_field()

# unknowns: only the c_ijk
Pc = PolynomialRing(Ka, [
    'c111','c112','c121','c122',
    'c211','c212','c221','c222'
], order='lex')
c111,c112,c121,c122,c211,c212,c221,c222 = Pc.gens()

C = {
    (1,1,1): c111, (1,1,2): c112,
    (1,2,1): c121, (1,2,2): c122,
    (2,1,1): c211, (2,1,2): c212,
    (2,2,1): c221, (2,2,2): c222
}

A = {
    (1,1,1): a111, (1,1,2): a112,
    (1,2,1): a121, (1,2,2): a122,
    (2,1,1): a211, (2,1,2): a212,
    (2,2,1): a221, (2,2,2): a222
}

def E(idx):
    # represents c_ijk + a_ijk*tau as the pair (constant term, tau coefficient)
    return (C[idx], A[idx])
    # We consider the case D=0 mod 4 here
def mul(x, y):
    u1, v1 = x
    u2, v2 = y
    return (u1*u2 + v1*v2*D/4, u1*v2 + u2*v1)
    #return (u1*u2 + v1*v2*(D-1)/4, u1*v2 + u2*v1+v1*v2)
    
eqs = []

    # Appending all of our equations
for i in (1,2):
    for ip in (1,2):
        for j in (1,2):
            for jp in (1,2):
                for k in (1,2):
                    for kp in (1,2):
                        lhs = mul(E((i,j,k)), E((ip,jp,kp)))
                        rhs = mul(E((ip,j,k)), E((i,jp,kp)))
                        eqs.append(lhs[0] - rhs[0])  # constant term
                        eqs.append(lhs[1] - rhs[1])
                        lhs = mul(E((i,j,k)), E((ip,jp,kp)))
                        rhs = mul(E((i,jp,k)), E((ip,j,kp)))
                        eqs.append(lhs[0] - rhs[0])  # constant term
                        eqs.append(lhs[1] - rhs[1])# tau term
                        lhs = mul(E((i,j,k)), E((ip,jp,kp)))
                        rhs = mul(E((i,j,kp)), E((ip,jp,k)))
                        eqs.append(lhs[0] - rhs[0])  # constant term
                        eqs.append(lhs[1] - rhs[1])# tau term
    # Using a Groebner basis to olve for c_ijk
I = ideal(eqs)
G = I.groebner_basis()
G
```

Last basis element gives quadratic equation for c222. We now solve for c222^2


DEFINING DISCRIMINANT



```python
a,b,c,d,e,f,g,h = var('a b c d e f g h')

Disc = (
    a^2*h^2 + b^2*g^2 + c^2*f^2 + d^2*e^2
    - 2*(a*b*g*h + c*d*e*f + a*c*f*h + b*d*e*g + a*d*e*h + b*c*f*g)
    + 4*(a*d*f*g + b*c*e*h)
)
subs_dict = {
    a: a111,
    b: a112,
    c: a121,
    d: a122,
    e: a211,
    f: a212,
    g: a221,
    h: a222
}

Disc_ijk = Disc.subs(subs_dict)
```

Substituting discriminant to simplify polynomial for c222


```python
g = expand(G[-1])
g = G[-1]

g2 = g.map_coefficients(lambda x: x.subs(D=Disc_ijk))
g2
```

Solve for c222


```python
A = g2.coefficient({c222: 2})
B = g2.coefficient({c222: 1})
C = g2.coefficient({c222: 0})

Delta = B^2 - 4*A*C

Delta
s = Delta.sqrt()

c222_sol1 =  (-B+s)/(2*A)
c222_sol2 = (-B-s)/(2*A)
c222_sol1
```

Find the other solutions of c\_ijk when we pick the negative square root:


```python
sol = c222_sol2
#change the line above to c222_sol1 for positive square root
G1 = [g.subs(c222=sol) for g in G]
G1 = G1[:-1] + [c222 - sol]
G1
```

Put solutions in a dictionary to refer to later


```python
sols = {}

for g in G1:
    for v in [c111, c112, c121, c122, c211, c212, c221, c222]:
        if v in g.variables():
            sols[v] = -g.subs({v: 0})
            break

sols
```

Picking alpha beta gamma in terms of a\_ijk and c\_ijk as in the project, we get the following norms: \(in the case D=0 mod 4, if we wanted to consider D=1 mod 4, the element norm changes\)


```python
D = var('D')
NI1 = c111*a211 - a111*c211
NI2 = (c111*a121 - a111*c121)/((c111^2)-((a111^2)*(D/4)))
NI3 = (c111*a112 - a111*c112)/((c111^2)-((a111^2)*(D/4)))

NI3 = NI3.subs(D=Disc_ijk)
NI2 = NI2.subs(D=Disc_ijk)

NI1=NI1.subs(sols)
NI2=NI2.subs(sols)
NI3=NI3.subs(sols)

N=NI1*NI2*NI3

N = N.simplify_full()
N
```
